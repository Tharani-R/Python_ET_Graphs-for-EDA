1. You will create a function named ‘Histogram’ which will take the entire dataset as input and
return the histograms for all the numerical variables in your dataset as .png files in your working
directory. Make sure that all the generated graphs have proper titles and axis labels.

```python
def Histogram(Data):                                             #Defining  function named histogram
    file = pd.read_csv(Data)                                     #Reading csv file
    col = file.select_dtypes(include=('int64','float64')).columns  #Getting index of all columns which are numerical type
    for i in col:                                                  #using for loop for all indexes in col variable
        plt.hist(file[i])                                          #Using hist function to plot histogram
        plt.title("Histogram for "+i)                              #Using title for plot                             
        plt.xlabel(i)                                              #Giving x label
        plt.ylabel("Frequency")                                    #Giving y label
        plt.savefig(i+"_Histogram")                                #saving figure
        plt.show()                                                 #It is used to show plot and show all figures
```

```python
import os
#changing the current directroy to the requried path
os.chdir(r"C:\Users\hp\OneDrive\Desktop\Python\End Term")
```

```python
#Function to create Histogram for all numerical variables in your dataset
def Histogram(Data):
    file = pd.read_csv(Data)
    
    #selecting only the numeric variables and saving those variable name in col
    col = file.select_dtypes(include=('int64','float64')).columns
    
    #for loop to plot histogram for each variable
    for i in col:
        plt.hist(file[i])   #plot histogram 
        plt.title("Histogram for "+i) #provides title to the plot
        plt.xlabel(i)   #provide x-axis label to the plot
        plt.ylabel("Frequency")    #provide y-axis label to the plot
        plt.savefig(i+"_Histogram")   #save the plot to current directory with the name provided
        plt.show()
```

```python
#Function to create Boxplot for all numerical variables in your dataset
def Boxplot(Data):
    file = pd.read_csv(Data)
    
    #selecting only the numeric variables and saving those variable name in col
    col = file.select_dtypes(include=('int64','float64')).columns
    
    #for loop to plot histogram for each variable
    for i in col:
        plt.boxplot(file[i],vert=False)    #plot Boxplot 
        plt.title("Boxplot for "+i)   #provides title to the plot
        plt.xlabel("Frequency")   #provide x-axis label to the plot
        plt.ylabel(i)    #provide y-axis label to the plot
        plt.savefig(i+"_Boxplot")    #save the plot to current directory with the name provided
        plt.show()  
```

```python
#Function to create Barplot for all categorical variables in your dataset
def Barchart(Data):
    file = pd.read_csv(Data)
    
    #selecting only the categorical variables and saving those variable name in col
    col = file.select_dtypes(include=('object')).columns
    #for loop to plot barplot for each variable
    for i in col:
        sns.countplot(file[i])  #plot Barplot 
        plt.title("Barchart for "+i)   #provides title to the plot
        plt.xlabel(i)   #provide x-axis label to the plot
        plt.ylabel("Frequency")    #provide y-axis label to the plot
        plt.savefig(i+"_Barchart")   #save the plot to current directory with the name provided
        plt.show()
```

```python
#Function to create graphs for all variables in your dataset and perform univariate analysis 
def Graphs(data):
    #calls the function Histogram
    Histogram(data)
    #calls the function Boxplot
    Boxplot(data)
    #calls the function Barchart
    Barchart(data)
```

```python
#Function to create Barplot for all categorical variables in your dataset
def Barchart(Data):
    
    #selecting only the categorical variables and saving those variable name in col
    col = Data.select_dtypes(include=('object')).columns
    #for loop to plot barplot for each variable
    for i in col:
        sns.countplot(Data[i])  #plot Barplot 
        plt.title("Barchart for "+i)   #provides title to the plot
        plt.xlabel(i)   #provide x-axis label to the plot
        plt.ylabel("Frequency")    #provide y-axis label to the plot
        plt.savefig(i+"_Barchart")   #save the plot to current directory with the name provided
        plt.show()
```

```python
#Function to create graphs for all variables in your dataset and perform univariate analysis 
def Graphs(data,dir,var=None):  #data-gets the path to the data  # dir- gets the path where the graph needs to be saved, 
                                #Var-to which variable the graphs should be plotted 
    os.chdir(dir)
    file = pd.read_csv(data)
    
    #if condition- if variables are specified is provided then index need to reduced by one as in python index starts at 1 and filter those variables and as data
    if var != None:
        var = list(var)
        for i in range(len(var)):
            var[i] = var[i] - 1
        print(var)
        data=file[file.columns[var]]
    
    #if variables are not specified then it takes the complete dataset
    else:
        data=file[file.columns]
        
    #calls the function Histogram       
    Histogram(data)
    #calls the function Boxplot
    Boxplot(data)
    #calls the function Barchart
    Barchart(data)
```

```python
import os
#changing the current directroy to the requried path
os.chdir(r"C:\Users\hp\OneDrive\Desktop\Python\End Term")
```

```python
#Function to create Histogram for all continuous variables in your dataset
def Histogram(Data,var):
    #for loop to plot histogram for each variable
    for i in var:
        plt.hist(Data[i])   #plot histogram 
        plt.title("Histogram for "+i) #provides title to the plot
        plt.xlabel(i)   #provide x-axis label to the plot
        plt.ylabel("Frequency")    #provide y-axis label to the plot
        plt.savefig(i+"_Histogram")   #save the plot to current directory with the name provided
        plt.show()
```

```python
#Function to create Boxplot for all continuous variables in your dataset
def Boxplot(Data,var):
    #for loop to plot Boxplot for each variable
    for i in var:
        plt.boxplot(Data[i],vert=False)    #plot Boxplot 
        plt.title("Boxplot for "+i)   #provides title to the plot
        plt.xlabel("Frequency")   #provide x-axis label to the plot
        plt.ylabel(i)    #provide y-axis label to the plot
        plt.savefig(i+"_Boxplot")    #save the plot to current directory with the name provided
        plt.show()  
```

```python
#Function to create Barplot for all categorical variables in your dataset
def Barchart(Data,var):
    #for loop to plot Barplot for each variable
    for i in var:
        sns.countplot(Data[i])  #plot Barplot 
        plt.title("Barchart for "+i)   #provides title to the plot
        plt.xlabel(i)   #provide x-axis label to the plot
        plt.ylabel("Frequency")    #provide y-axis label to the plot
        plt.savefig(i+"_Barchart")   #save the plot to current directory with the name provided
        plt.show()
```

```python
#Function to create Barplot between each continuous variables vs each categorical variables in your dataset
def Barplot_bivariate(data,cont,cat):
    #for loop to pick each variable categorical variable
    for i in cat:
        #for loop to pick each variable continuous variable
        for j in cont:
            plt.bar(data[i],data[j]) #plot Barplot 
            plt.title("Barchart for "+i+"vs"+j)   #provides title to the plot
            plt.xlabel(i)   #provide x-axis label to the plot
            plt.ylabel(j)   #provide y-axis label to the plot
            plt.savefig(i+" vs "+j+"_Barchart")   #save the plot to current directory with the name provided
            plt.show()
```

```python
#Function to create Barplot between each continuous variables in your dataset
def Scatterplot_bivariate(data,cont):
    #for loop to pick each variable continuous variable
    for i in cont:
        #for loop to pick each variable continuous variable
        for j in cont:
            plt.scatter(data[i],data[j])  #plot Scatterplot
            plt.title("Scatter plot for "+i+"vs"+j)  #provides title to the plot
            plt.xlabel(i)   #provide x-axis label to the plot
            plt.ylabel(j)   #provide y-axis label to the plot
            plt.savefig(i+"vs"+j+"_Scatter plot")  #save the plot to current directory with the name provided
            plt.show()
```

```python
#Function to create Boxplot between each continuous variables vs each categorical variables in your dataset
def Boxplot_bivariate(data,cont,cat):
    #for loop to pick each variable categorical variable
    for i in cat:
        #for loop to pick each variable continuous variable
        for j in cont:
            data.boxplot(column=j , by=i, vert = False)  #plot Scatterplot
            plt.title("Boxplot for "+i+"vs"+j)   #provides title to the plot
            plt.xlabel(j)   #provide x-axis label to the plot
            plt.ylabel(i)   #provide y-axis label to the plot
            plt.savefig(i+"vs"+j+"_Boxplot")  #save the plot to current directory with the name provided
            plt.show()
```

```python
#Function to create graphs for all variables in your dataset and perform univariate analysis 
def Graphs(data,dir,var=None):  #data-gets the path to the data  # dir- gets the path where the graph needs to be saved, 
                                #Var-to which variable the graphs should be plotted
    os.chdir(dir)
    file = pd.read_csv(data)
    cont = []
    cat = []
    #if condition- if variables are specified is provided then index need to reduced by one as in python index starts at 1 and filter those variables and as data
    if var != None:
        var = list(var)
        for i in range(len(var)):
            var[i] = var[i] - 1       
        print(var)
        data=file[file.columns[var]]
    #if variables are not specified then it takes the complete dataset
    else:
        data=file[file.columns]
    
    #some cases some variable looks like numeric but it will actully be categorical, so seperating those variables add those into categorical
    col = data.select_dtypes(include=('int64','float64')).columns
    for i in col:
        unique = len(file[i].unique())
        if unique >= 8:
            cont.append(i)
        else:
            cat.append(i)
    col = (data.select_dtypes(include=('object')).columns)
    for i in col:
        cat.append(i)
    print(cont,cat)
    #calls the function Histogram      
    Histogram(data,cont)
    #calls the function Boxplot  
    Boxplot(data,cont)
    #calls the function Barchart  
    Barchart(data,cat)
    #calls the function Barplot_bivariate  
    Barplot_bivariate(data,cont,cat)
    #calls the function Scatterplot_bivariate  
    Scatterplot_bivariate(data,cont)
    #calls the function Boxplot_bivariate  
    Boxplot_bivariate(data,cont,cat)
```
